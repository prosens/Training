```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
pop = pd.read_csv('DataSets/population.csv')
pop.head()
```

```python
px.line(pop.query('Year > 1900'), x='Year', y='Total population (Gapminder, HYDE & UN)', color='Entity')
```

```python
pop = pop.rename(columns={'Total population (Gapminder, HYDE & UN)':'population'})
```

```python
gdp = pd.read_csv('DataSets/gdp.csv')
gdp
```

```python
pop_2017 = pop.query('Year == 2017')
gdp_2017 = gdp.query('Year == 2017')
```

```python
pop_gdp_2017 = pd.merge(left=gdp_2017, right=pop_2017, left_on='Entity', right_on='Entity', how='outer')
pop_gdp_2017
```

```python
pop_gdp_2017['gdp_per_capita'] = pop_gdp_2017['GDP (constant 2010 US$)'] / pop_gdp_2017['population']
pop_gdp_2017.sort_values(by='gdp_per_capita', ascending=True)
```

```python
px.bar(pop_gdp_2017.sort_values(by='gdp_per_capita', ascending=True), x='Entity', y='gdp_per_capita')
```

```python
gdp_pop = pd.merge(left=gdp, right=pop, on=['Entity', 'Year'], how='left')
gdp_pop
```

```python
gdp_pop['gdp_per_capita'] = gdp_pop['GDP (constant 2010 US$)'] / gdp_pop['population']
gdp_pop = gdp_pop.set_index('Entity')
gdp_pop
```

```python
gdp_pop
```

```python
gdp_pop_1960 = gdp_pop.query('Year == 1960')
gdp_pop_1960
```

```python
gdp_pop['gdp_per_capita_ratio'] = gdp_pop['gdp_per_capita'] / gdp_pop_1960['gdp_per_capita']
gdp_pop = gdp_pop.dropna()
```

```python
gdp_pop = gdp_pop.reset_index()
```

```python
countries = ['India', 'China', 'Japan', 'United Kingdom']
px.line(gdp_pop.query('Entity in @countries'), x = 'Year', y = 'gdp_per_capita_ratio', color='Entity')
```
