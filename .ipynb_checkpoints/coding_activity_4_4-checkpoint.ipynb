### Histograms, KDE Plots, Violin Plots, Box Plots

This activity focuses on producing the plots demonstrated in Video 4.6 and Video 4.7.  Specifically, histograms, Kernel Density Estimate plots, Violin, and Box plots.  The dataset is the `gapminder` data from plotly.

```python
import pandas as pd
import plotly.express as px
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Saving the Plots

For each problem you will be asked to produce a specific plot and save that image to the `images` folder with the appropriate title.  

```python
gapminder = px.data.gapminder()
gapminder
```

### Problem 1: Histogram of Life Expectancy with `plotly`

Create a histogram of the Life Expectancy data using `plotly` and save the plot to `ans1.png` in the `images` folder.

```python
#make a histogram of life expectancy in plotly
hist = px.histogram(gapminder, x= 'lifeExp')
hist
```

### Problem 2: Histogram of Life Expectancy with `seaborn`

Create a histogram of the Life Expectancy data using `seaborn` and save the plot to `ans2.png` in the `images` folder.

```python
#make a histogram in seaborn
sns.histplot(data=gapminder, x="lifeExp")
```

### Problem 3: Histogram and KDE of Life Expectancy with `seaborn` 

Create a histogram with the KDE of the Life Expectancy data using `seaborn` and save the plot to `ans3.png` in the `images` folder.

```python
#make a kde plot in seaborn
sns.displot(data=gapminder, x="lifeExp", kde=True)
```

### Problem 4: Histogram, KDE, and rugplot of Life Expectancy with `seaborn`

Create a histogram and KDE of the Life Expectancy data along with the rugplot using `seaborn` and save the plot to `ans4.png` in the `images` folder.

```python
#make a rug plot with seaborn
sns.displot(data=gapminder, x="lifeExp", kde=True)
sns.rugplot(data=gapminder, x="lifeExp")
```

### Problem 5: Violin Plot of Life Expectancy with `plotly`

Create a Violin Plot of the Life Expectancy data using `plotly` and save the plot to `ans5.png` in the `images` folder.

```python
#create a violin plot with plotly
fig = px.violin(gapminder, y='lifeExp')
fig
```

### Problem 6: Violin Plot with points of Life Expectancy with `plotly`

Create a Violin Plot of the Life Expectancy data alongside the actual points using `plotly` and save the plot to `ans6.png` in the `images` folder.

```python
#violin in plotly with points = "all"
fig = px.violin(gapminder, y='lifeExp',points = 'all')
fig
```

### Problem 7: Violin Plot of Life Expectancy by Continent with `plotly`

Create a Violin Plot of the Life Expectancy by continent using `plotly` and save the plot to `ans7.png` in the `images` folder.

```python
#violin plot by continent in plotly
fig = px.violin(gapminder, x='continent',y='lifeExp')
fig
```

### Problem 8: Violin Plot of Life Expectancy by Continent with `seaborn`

Create a Violin Plot of the Life Expectancy by continent using `seaborn` and save the plot to `ans8.png` in the `images` folder. Note that you will set your `x` argument to continent and `y` to lifeExp.

```python
#violin plot by continent in seaborn
sns.violinplot(x="continent", y="lifeExp", data=gapminder)
```

### Problem 9: Box Plot of Life Expectancy by Continent with `plotly`

Create a Box Plot of the Life Expectancy by continent using `plotly` and save the plot to `ans9.png` in the `images` folder. 

```python
#boxplot by continent with plotly with y = lifeExp
fig = px.box(gapminder,x="continent", y="lifeExp")
fig
```

### Problem 10: Box Plot of Life Expectancy by Continent with `seaborn`

Create a Box Plot of the Life Expectancy by continent using `seaborn` and save the plot to `ans10.png` in the `images` folder. Note that you will set your `x` argument to continent and `y` to lifeExp.

```python
#seaborn boxplot by continent
sns.boxplot(x="continent", y="lifeExp", data=gapminder)
```

### Problem 11: Scatterplot of Life Expectancy vs. GDP including marginal histograms with `plotly`

Create a Scatterplot Life Expectancy vs. GDP including marginals as histograms using `plotly`. Save the plot to `ans11.png` in the `images` folder. 

```python
#plotly gdp v life exp marginal_x = "histogram" and y
fig = px.scatter(gapminder, x="lifeExp", y="gdpPercap", marginal_x="histogram", marginal_y="histogram")
fig
```

### Problem 12: Scatterplot of Life Expectancy vs. GDP including marginal box plots with `plotly`

Create a Scatterplot Life Expectancy vs. GDP including marginals as box plots using `plotly`. Save the plot to `ans12.png` in the `images` folder. 

```python
#marginals with boxplot
fig = px.scatter(gapminder, x="lifeExp", y="gdpPercap", marginal_x="box", marginal_y="box")
fig
```

### Problem 13: Density Heatmap of Life Expectancy vs. GDP including marginal histograms with `plotly`

Create a Density Heatmap of Life Expectancy vs. GDP including marginals as histograms using `plotly`. Save the plot to `ans13.png` in the `images` folder. 

```python
#density heatmap with plotly
fig = px.density_heatmap(gapminder, x="lifeExp", y="gdpPercap", marginal_x="histogram", marginal_y="histogram")
fig
```

### Problem 14: `jointplot` of Life Expectancy vs. GDP with `kind = hex` using `seaborn`

Create a `jointplot` of Life Expectancy vs. GDP and set the `kind` argument to `hex`. Save the plot to `ans14.png` in the `images` folder. 

```python
#seaborn jointplot with kind = "hex"
sns.jointplot(data=gapminder, x="lifeExp", y="gdpPercap", kind="hex")
```
