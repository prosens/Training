```python
from scipy.stats import uniform
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dist1 = uniform(loc=10, scale=5)
dist1.pdf(10)
```

```python
upoints = np.linspace(7,18,200)
ppoints = dist1.pdf(upoints)

plt.plot(upoints,ppoints)
```
