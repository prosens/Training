```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.max_rows = 10
```

```python
df = pd.read_csv('gdp.csv')
```

```python
df.sample(5)
```

```python
df['gdp'] = df['GDP (constant 2010 US$)'] / 1_000_000_000
```

```python
df
```

```python
df[df['Entity'] == 'China']
```

```python
(df['Entity'] == 'China').iloc[1775:1790]
```

```python
df[(df['Entity'] == 'China') & (df['Year'] == 2017)]
```

```python
df.query('Entity == "China" and Year == 2017')
```

```python
list_of_countries = ["China", "United States", "India", "Japan", "Germany", "United Kingdom"]
top6 = df[df['Entity'].isin(list_of_countries)]
top6.sample(10)
```

```python
top6 = df.query('Entity in @list_of_countries')
top6.sample(10)
```

```python
top6.plot(kind = 'bar', x='Entity', y='gdp')
```

```python
sns.barplot(data=top6, x='Entity', y='gdp')
plt.xticks(rotation = 90)
```
