```python
from scipy.stats import uniform
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dist = norm(loc = 5, scale = 2)
dist1 = dist.rvs(100, random_state = 12)
sample = np.linspace(dist.mean() - 3* dist.std(), dist.mean() + 3* dist.std(), 1000) 
plt.plot(sample, dist.pdf(sample), color = 'black', linewidth = 4, label = 'distribution')
plt.hist(dist1, density=True, alpha = 0.2, bins = 10, edgecolor = 'black', label = 'dist')
    
```
